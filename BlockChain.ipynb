```python
import hashlib

class MerkleTree:
    def __init__(self, transactions):
        self.transactions = transactions
        self.tree = self.build_tree()

    def build_tree(self):
        if len(self.transactions) % 2 != 0:
            self.transactions.append(self.transactions[-1])

        tree = [None] * (2 * len(self.transactions) - 1)

        # Rellena las hojas del árbol con los hashes de transacción.
        for i in range(len(self.transactions)):
            tree[i + len(self.transactions) - 1] = hashlib.sha256(self.transactions[i].encode()).hexdigest()

        # Completa el resto del árbol.
        for i in range(len(self.transactions) - 2, -1, -1):
            tree[i] = hashlib.sha256((tree[2*i + 1] + tree[2*i + 2]).encode()).hexdigest()

        return tree

    def get_root(self):
        return self.tree[0]


```

```python
import hashlib

class Block:

  def __init__(self,previous_block,data):
    self.previous_block=previous_block
    self.data=data
    self.nonce=0
    self.merkle_root = ""


  def hashCreator(self):
    info = self.merkle_root + self.data + self.previous_block + str(self.nonce)
    hash=hashlib.sha256(info.encode()).hexdigest()
    return hash

  def set_merkle_root(self, merkle_root):
        self.merkle_root = merkle_root

  def print_info(self):
      print("Nonce:", self.nonce)
      print("Merkle Root:", self.merkle_root)
      print("Previous Block Hash:", self.previous_block)
      print("Data:", self.data)
      print("\n")

  def verify_transaction(self, block_chain):
          # Verifica si la longitud de la cadena de datos de la transacción es mayor que cero
          if len(self.data) == 0:
              return False

          # Verifica si la transacción ya existe en el bloque actual
          for block in block_chain.chain:
              if block.data == self.data:
                  return False

          return True




class Blockchain:
  dificulty = 4

  def __init__(self):
    self.chain= []

  def add(self,block):
    self.chain.append(block)

  def mine(self,block):
    global contador
    if not self.chain:
            block.set_merkle_root("")
    else:
            block.set_merkle_root(self.calculate_merkle_root())

    while True:
      hash_value = block.hashCreator()
      if hash_value[:self.dificulty] == "0"*self.dificulty:
        self.add(block)
        contador+=1
        break
      else:
        block.nonce +=1

  def calculate_merkle_root(self):
        transactions = [block.data for block in self.chain]
        merkle_tree = MerkleTree(transactions)
        return merkle_tree.get_root()


  def print_chain(self):
      for block in self.chain:
          block.print_info()
```
